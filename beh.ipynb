```python
import os
import pandas as pd
import numpy as np
```

```python
beh = pd.read_csv('./train/训练数据集_beh.csv')

beh['day'] = beh['page_tm'].apply(lambda x: x.split(' ')[0])
beh['time'] = beh['page_tm'].apply(lambda x: x.split(' ')[1])
del beh['page_tm']
```

```python
info = beh.groupby(['id', 'page_no', 'day']).agg({'flag': 'count'})
one_id = info[info.index.get_level_values('id') == 'U10001']
```
